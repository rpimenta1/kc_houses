### 0. Importando bibliotecas

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('datasets/kc_house_data.csv')
pd.set_option('display.float_format', lambda x: '%.2f' % x) #definindo as casas decimais
data.head()
```

### 1. Qual o número de imoveis por ano de construção?


```python
data[['id','yr_built']].groupby('yr_built').count()
```

### 2. Qual o menor número de quartos por ano de construção de imóveis?


```python
data[['bedrooms','yr_built']].groupby('yr_built').min()
```

### 3. Qual o preço de compra mais alto por cada número de quartos?


```python
data[['price','bedrooms']].groupby('bedrooms').max()
```

### 4. Qual a soma de todos os preços de compra por número de quartos?


```python
data[['price','bedrooms']].groupby('bedrooms').sum()
```

### 5. Qual a soma de todos os preços de compra por número de quartos e banheiros?


```python
data[['price','bedrooms','bathrooms']].groupby(['bedrooms','bathrooms']).sum()
```

### 6.  Qual o tamanho médio de salas dos imóveis por ano de construção?


```python
data[['sqft_living','yr_built']].groupby('yr_built').mean()
```

### 7. Qual o tamanho mediano das salas dos imóveis por ano de construção?


```python
data[['sqft_living','yr_built']].groupby('yr_built').median()
```

### 8. Qual o desvio padrão do tamanho das salas dos imóveis por ano de construção?


```python
data[['sqft_living','yr_built']].groupby('yr_built').std()
```

### 9. Como é o crescimento total de preços de compras dos imoveis por ano, por dia e pela semana do ano?


```python
# Verificar o crescimento total de preços de compra dos imóveis por ano
# Criando gráfico de barras que contenha 'anos' no eixo x e 'Soma de preços no eixo y'
# Colocar 'date' como tipo data e criar coluna com o ano de publicação

data['year'] = pd.to_datetime(data['date']).dt.year
by_year = data[['price', 'year']].groupby('year').sum().reset_index()

plt.figure(figsize =(20,12))
plt.bar(by_year['year'],by_year['price'])
```

```python
# Criando gráfico de linha que contenha 'dias' no eixo x e 'média de preços no eixo y'

data['day'] = pd.to_datetime(data['date'])
by_day = data[['price', 'day']].groupby('day').mean().reset_index()

plt.figure(figsize =(20,12))
plt.plot(by_day['day'],by_day['price'])
```

```python
# Criando gráfico de linha que contenha 'semanas' no eixo x e 'média de preços no eixo y'

data['semana'] = pd.to_datetime(data['date']).dt.strftime('%Y-%U')
by_semana_ano = data[['price', 'semana']].groupby('semana').mean().reset_index()

plt.figure(figsize =(20,12))
plt.plot(by_semana_ano['semana'],by_semana_ano['price'])
plt.xticks(rotation = 60);
```
